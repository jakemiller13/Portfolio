**Goal: Using data from boston.gov, I explore whether giving a restaurant an inspection violation has an impact on the way the restaurant operates in the future.**

This data is collected from [Analyze Boston](https://data.boston.gov/), which offers many dataset related to the city. It also offers a very convenient API to access, which really helps the data gathering process.
I had no idea when I approached this data whether giving a restaurant an inspection violation actually had any impact on the functioning of the restaurant, but I think we get a pretty good picture by the end.

This specific dataset can be accessed here: 
https://data.boston.gov/dataset/food-establishment-inspections

```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
from urllib.request import urlopen
from urllib.error import HTTPError
import matplotlib.pyplot as plt
import seaborn as sns
```

This is the link to access the API. We will cycle through this later to pull in all the data offered on the website:

```python
link = 'https://data.boston.gov/datastore/odata3.0/'\
        + '4582bec6-2b4f-4f9e-bc55-cbaa73117f4c'
```

We create an empty dataframe to use to collect the data:

```python
columns = ['id',
           'businessname',
           'dbaname',
           'legalowner',
           'namelast',
           'namefirst',
           'licenseno',
           'issdttm',
           'expdttm',
           'licstatus',
           'licensecat',
           'descript',
           'result',
           'resultdttm',
           'violation',
           'viollevel',
           'violdesc',
           'violdttm',
           'violstatus',
           'statusdate',
           'comments',
           'address',
           'city',
           'state',
           'zip',
           'property_id',
           'location']
df = pd.DataFrame(columns = columns)
```

The first function here returns a link as a string. Using this, we increment the API count by 1000 each iteration to ensure we don't timeout on the website. Since this dataset is pretty large, we do occassionally have issues accessing the website, so we give the option of starting the increment at something other than 0.

```python
def inc_address(link, inc):
    '''
    Retrieve 1000 rows at a time to avoid timing out
    '''
    increment = inc * 1000
    link = link + '?$top=1000&$skip=' + str(increment)
    return link
```

This second function returns a temporary dataframe of the 1000 lines of the API we access. This makes it very easy to append to the larger dataframe each iteration.

```python
def create_dataframe(tree, start, columns):
    '''
    Used in get_data()
    Returns dataframe created from *tree*
    Must supply *start* which is root level at which info starts, e.g. "4"
    *columns* are column headers for dataframe
    '''
    info_list = []
    while True:
        entry = []
        try:
            for child in tree[start][-1][-1]:
                entry.append(child.text)
            info_list.append(entry)
            start += 1
        except IndexError:
            break
    return pd.DataFrame(info_list, columns = columns)
```

This second function is where we actually collect the data. Again, if we have any issues accessing or staying connected to the API, we return the incomplete dataframe, which allows us to just pick up where we left off.

```python
def get_data(df, link):
    '''
    Returns data from link as ElementTree.Element
    Occasionally runs into HTTPError mid-execution
    If that happens, returns df as-is. To continue running, re-execute this
        function with correct increment fed into inc_address to get
        starting link address
    '''
    count = 0
    while True: # uncomment when running full, otherwise = 25
        try:
            if count % 10 == 0:
                print('Count: ' + str(count) + ' | df shape: ' + str(df.shape))
            link = inc_address(link, count)
            response = urlopen(link).read().decode('utf-8')
            tree = ET.fromstring(response)
            if tree[-1].attrib == {}:
                temp_df = create_dataframe(tree, start = 4, columns = columns)
                df = pd.concat([df, temp_df], ignore_index = True)
                break
            else:
                temp_df = create_dataframe(tree, start = 4, columns = columns)
                df = pd.concat([df, temp_df], ignore_index = True)
                count += 1
        except HTTPError:
            return df
    return df

df = get_data(df, link)
```

Ultimately, we don't really care about the specific times that violations were given, so let's split up the datetime columns in separate date and time columns. There are some instances which are missing information in these columns (`np.nan`), which can cause `IndexError`, so we need to make sure these don't interrupt our `for` loop.

```python
def split_datetime(df, dttm_col):
    '''
    Splits datetime column into 2 separate date & time columns
    Returns [date, time]
    '''    
    date, time = [], []
    dttm_split = df[dttm_col].apply(lambda x: x.split())
    for i in dttm_split:
        try:
            date.append(i[0])
            time.append(i[1])
        except IndexError:
            date.append(np.nan)
            time.append(np.nan)
    return date, time
```

```python
def add_dttm_cols(df, dttm_col, date, time):
    '''
    Adds new columns in df for *date* and *time*
    *date* and *time* are from split_datetime()
    *dttm_col* is the string name of column being split
    Returns new column names, used when converting to pd.datetime
    '''
    idx = list(df.columns).index(dttm_col)
    col_name = dttm_col[:-4]
    date_name = col_name + '_date'
    time_name = col_name + '_time'
    df.insert(idx + 1, date_name, date)
    df.insert(idx + 2, time_name, time)
    return date_name, time_name
```

```python
datetime_cols = ['issdttm', 'expdttm', 'resultdttm', 'violdttm']
date_cols = []
time_cols = []

for col in datetime_cols:
    date, time = split_datetime(df, col)
    date_name, time_name = add_dttm_cols(df, col, date, time)
    date_cols.append(date_name)
    time_cols.append(time_name)
```

Now we cast these columns to `pd.Timestampe` type.

```python
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors = 'ignore')
```

There are also some columns which seem to fluctuate in datatype, so we cast all instances to `str` that should be.

```python
str_cols = ['viollevel', 'violstatus', 'violdesc']
df[str_cols] = df[str_cols].astype(str)
```

One thing that's critical to note about this dataset is that it doesn't *only* incorporate "Failing" inspections. It also captures "Passing" inspections, which can be seen below, which aren't helpful in this analysis.

```python
print('Results of inspections: \n' + str(np.unique(df.result)))
```

Let's get rid anything related to "Passing" inspections. Note that we aren't making an entirely new dataframe, but instead just making a slice of the "Failing" instances.

```python
df_fail = df[(df.result != 'HE_Pass') &
             (df.result != 'Pass') &
             (df.result != 'PassViol')]
```

This function returns the unique dates of a specific "date_col", and the number of instances of that date. This gives us a count of how many times something happened (e.g. how many violations) on a specific date.

```python
def date_counts(df, date_col):
    '''
    Returns dates and counts of *date_col*
    Used when counting violations given
    Temporarily drops rows that are NaT to avoid extra counting
    '''
    return np.unique(df.dropna(subset = [date_col])[date_col],
                     return_counts = True)
```

```python
viol_date, viol_counts = date_counts(df_fail, 'viol_date')
```

Ok, now let's see when all of the violations happened:

```python
plt.plot(viol_date, viol_counts)
plt.ylim(0, max(viol_counts) + 10)
plt.xlabel('Date Found')
plt.ylabel('Number of Violations Found')
plt.title('Total Violations Found')
plt.grid(b = True, axis = 'y', linestyle = '--')
plt.show()
```

Well that plot is not actually that useful. It shows us that inspections and violations have more or less happend with the same frequency since 2008, which isn't that useful for what we are looking at. Let's try something else.

```python
years = np.unique(pd.DatetimeIndex(viol_date).year)
n, bins, patches = plt.hist(pd.DatetimeIndex(viol_date).year,
                            bins = np.arange(min(years), max(years) + 2) - 0.5,
                            edgecolor = 'black',
                            linewidth = 1)
cm = plt.cm.YlOrRd
for i, p in enumerate(patches):
    plt.setp(p, 'facecolor', cm(i/len(years)))
plt.ylabel('Number of Violations Found')
plt.title('Violations Issued By Year')
plt.grid(b = True, axis = 'y', linestyle = '--', color = 'k', alpha = 0.2)
plt.show()
```

Ok, that's a little more useful. It shows that violations were issued at roughly the same amount each year (note that this analysis was done in mid-2019), so we can assume that when we start to look at how violations impact restaurants, there won't be any skew due to anomolous years.

Let's looks at days when the most violations were found. Once again, we can see that these are pretty evenly spread out over the duration of the dataset:

```python
top_10_index = viol_counts.argsort()[-10:][::-1]
top_10_dates = viol_date[top_10_index]
top_10_counts = viol_counts[top_10_index]
print('Top 10 days violations were identified [number identified]:')
for i, j in zip(pd.DatetimeIndex(top_10_dates), top_10_counts):
    print(str(i.date()) + ' [' + str(j) + ']')
```

Let's now take a look at what the most common types of violations are:

```python
viol_types, viol_type_counts = np.unique(df_fail.violdesc,
                                         return_counts = True)
top_viol_index = viol_type_counts.argsort()[-20:][::-1]
top_viols = viol_types[top_viol_index]
top_viols_counts = viol_type_counts[top_viol_index]

ax = sns.barplot(x = top_viols_counts, y = top_viols)
ax.set_title('Top Violations')
ax.grid(True, axis = 'x')
plt.show()
```

The following function finds the restaurants with the most violations each year. Note that we are using the license number of the restaurant here, not the name. This avoids completely different "Dunkin' Donuts" establishments from being counted as one restaurant.

```python
def find_top_violators(df, year, k_violators):
    '''
    Finds top *k-violators* based on *year*
    *df*: can be either dataframe, but should be "df_fail"
    returns: top_names, top_lic, top_counts
    '''
    rows = [row for row in range(len(df))
            if str(year) in str(df.iloc[row].viol_date)]
    df_year = df.iloc[rows]
    lic_no, lic_counts = np.unique(df_year['licenseno'], return_counts = True)
    top_index = lic_counts.argsort()[-k_violators:][::-1]
    top_counts = lic_counts[top_index]
    top_lic = lic_no[top_index]
    top_names = [df[df.licenseno == i]['businessname'].iloc[0].upper()
                 for i in top_lic]
    return top_names, top_lic, top_counts
```

This function will be used to repeatedly plot the top violators for a series of years, so we can see how violations impact restaurants over time.

```python
def plot_top_violators(df, year, k_violators):
    '''
    Plots top violators by accessing "find_top_violators"
    *df*: can be either dataframe, but should be "df_fail"
    *year*: year to look at top violators
    *k_violators*: top k-violators for year
    '''
    top_names, top_lic, top_counts = find_top_violators(df, year, k_violators)
    plt.figure(figsize = (10, 10))
    for i, lic in enumerate(top_lic):
        temp_years, temp_counts = np.unique(pd.DatetimeIndex(
                                  df[df['licenseno'] == lic]
                                  ['viol_date']).year,
                                  return_counts = True)
        plt.plot(temp_years, temp_counts, label = top_names[i])
    plt.legend(loc = 'best')
    plt.xlim(2008, 2018)
    plt.xlabel('Year')
    plt.ylabel('Violations')
    plt.title('Violations Per Year for Top ' + str(year) + '-Violators')
    plt.show()
```

First, we will take a look at the restaurants with the most violations last year, just as a personal investigation to see how many of them I've already eaten at...

```python
top_2018_names, top_2018_lic, top_2018_counts = \
                              find_top_violators(df_fail, 2018, 20)
ax = sns.barplot(x = top_2018_counts, y = top_2018_names)
ax.set_title('Violations Found in 2018')
ax.grid(True, axis = 'x')
plt.show()
```

Now, let's look at the restaurants with the most violations for the duration of the entire dataset. By looking at 3 different points in time, we can see how issuing violations impacts a restaurant.
* 2008: We can see that by issuing violations here, these restaurants seem to keep other violations to a minimum for the rest of the dataset
* 2013: By taking a date right in the middle of the dataset, we can see how restaurants were doing both before and after getting many violations
* 2018: And by taking a date at the very end of the dataset, we can get a good idea of how restaurants were doing up to the most recent point in time, before getting a bunch of violations.

```python
plot_top_violators(df_fail, 2008, 20)
plot_top_violators(df_fail, 2013, 20)
plot_top_violators(df_fail, 2018, 20)
```

By taking these plots all together, it appears that serving violations to restaurants actually does have some impact on how they operate afterwards. The first plot shows that after issuing violations, most restaurants receive less violations going forward. There is a very interesting jump in 2014 which would be worth investigating in future work. It's a little difficult to tell much from the second plot - restaurants receive similar levels of violations before and after the peak year of 2013. But the final plot (Top 2018-Violators) is very interesting. You can see a clear upward trend over the decade before all of these restaurants received many violations last year. It would be very interesting to do this analysis again in 3-5 years and see what impact 2018 had on them.
